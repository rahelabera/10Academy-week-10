```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("data/processed/enriched_ethiopia_fi_unified_data.csv")

impact = df[df['record_type'] == 'impact_link'].copy()

# Optional: map magnitude to numbers (low=1, medium=2, high=3)
impact['impact_num'] = impact['impact_magnitude'].map({'low':1, 'medium':2, 'high':3})

pivot = impact.pivot_table(
    index='event_description',          # or use event name/date
    columns='related_indicator',
    values='impact_num',
    aggfunc='first'
)

sns.heatmap(pivot, annot=True, cmap='YlGnBu', fmt='.0f')
plt.title("Event Impact Association Matrix")
plt.savefig("reports/figures/event_impact_heatmap.png")
```
